### PART A - THE EFFECT OF DELAYING TREATMENT ON PATIENT DISEASE

```python
# 6.00.2x Problem Set 4

import numpy
import random
import pylab
from ps3b import *
%matplotlib inline

#
# PROBLEM 1
#        
def simulationDelayedTreatment(numTrials):
    """
    Runs simulations and make histograms for problem 1.

    Runs numTrials simulations to show the relationship between delayed
    treatment and patient outcome using a histogram.

    Histograms of final total virus populations are displayed for delays of 300,
    150, 75, 0 timesteps (followed by an additional 150 timesteps of
    simulation).

    numTrials: number of simulation runs to execute (an integer)
    """
    
    # TODO

    #parameters
    numViruses = 100
    maxPop = 1000
    maxBirthProb = 0.1
    clearProb = 0.05
    resistances = {'guttagonol': False}
    mutProb = 0.005

    #Monte Carlo Simulation
    treatOnSteps = [300, 150, 75, 0]#delays
    plot = 1
    for treatOnStep in treatOnSteps:
        steps = treatOnStep + 150
        finalVirusPop = []
        for _ in range(numTrials):
            viruses = [ResistantVirus(maxBirthProb, clearProb, resistances, mutProb) for j in range(numViruses)]
            patient = TreatedPatient(viruses, maxPop)
            for step in range(steps):
                if step == treatOnStep:
                    patient.addPrescription("guttagonol")
                patient.update()
            finalVirusPop.append(patient.getTotalPop())
        pylab.subplot(4, 1, plot)
        pylab.hist(finalVirusPop, bins=10)
        plot += 1
        pylab.title("Final virus population simulation")
        pylab.xlabel("Population")
        pylab.ylabel("# trials")
        pylab.legend()
        pylab.show()
        
        #Problem 1
        lessThan50 = [pop for pop in finalVirusPop if pop <= 50]
        print len(lessThan50) / float(len(finalVirusPop))

simulationDelayedTreatment(1000)
```

### PART B - DESIGNING A TREATMENT PLAN WITH TWO DRUGS

```python
import numpy
import random
import pylab
from ps3b import *
%matplotlib inline

#Problem 1, 2

def simulationTwoDrugsDelayedTreatment(numTrials):
    """
    Runs simulations and make histograms for problem 2.

    Runs numTrials simulations to show the relationship between administration
    of multiple drugs and patient outcome.

    Histograms of final total virus populations are displayed for lag times of
    300, 150, 75, 0 timesteps between adding drugs (followed by an additional
    150 timesteps of simulation).

    numTrials: number of simulation runs to execute (an integer)
    """
    # TODO
    numViruses = 100
    maxPop = 1000
    maxBirthProb = 0.1
    clearProb = 0.05
    resistances = {'guttagonol': False, 'grimpex': False}
    mutProb = 0.005

    #Monte Carlo Simulation
    plot = 1
    treatOnSteps = [300, 150, 75, 0]#delays
    
    for treatOnStep in treatOnSteps:
        steps = treatOnStep + 150 + 150
        finalVirusPop = []
        for _ in range(numTrials):
            viruses = [ResistantVirus(maxBirthProb, clearProb, resistances, mutProb) for j in range(numViruses)]
            patient = TreatedPatient(viruses, maxPop)
            for step in range(steps):
                if step == 150:
                    patient.addPrescription("guttagonol")
                if step == 150 + treatOnStep:
                    patient.addPrescription("grimpex")
                patient.update()
            finalVirusPop.append(patient.getTotalPop())
        pylab.subplot(4, 1, plot)
        pylab.hist(finalVirusPop, bins=10)
        plot += 1
        pylab.title("Final virus population simulation")
        pylab.xlabel("Population")
        pylab.ylabel("# trials")
        pylab.legend()
        pylab.show()
        
        lessThan50 = len([v for v in finalVirusPop if v <= 50])
        #print "Percentage of patients were cured: %.3f \n" % lessThan50 / float(numTrials)
        print "Percentage of patients were cured: ", lessThan50 / float(numTrials)
        
simulationTwoDrugsDelayedTreatment(1000)
```
